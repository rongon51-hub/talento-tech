```python
Tablero = [" "]*9

print("  0 | 1 | 2  ")
print("------------")
print("  3 | 4 | 5  ")
print("------------")
print("  6 | 7 | 8  ")
print("=============")
print()


def mostrar_tablero(t):
    print(f"   {t[0]} |  {t[1]} | {t[2]} ")
    print("------------")
    print(f"   {t[3]} |  {t[4]} | {t[5]} ")
    print("------------")
    print(f"   {t[6]} |  {t[7]} | {t[8]} ")


def llamar_funciones():
    # This function was empty and caused an IndentationError.
    # It's now implemented to display the initial board.
    mostrar_tablero(Tablero)

# The second 'def mostrar_tablero(Tablero):' was a redundant definition
# and also caused an IndentationError. It has been removed.

def pedir_jugada(simbolo_jugador):
    while True: # Se repite hasta que la jugada sea exitosa
        casilla_str = input(f"Jugador {simbolo_jugador}, selecciona casilla (0-8): ")
        try:
            casilla = int(casilla_str)
            if 0 <= casilla <= 8:
                if Tablero[casilla] == " ":
                    Tablero[casilla] = simbolo_jugador
                    break # Jugada exitosa, salimos del bucle
                else:
                    print("Esa casilla ya está llena. Intenta con otra.")
            else:
                print("Número fuera de rango (0-8).")
        except ValueError:
            print("Por favor, escribe un número válido.")


def verificar_ganador(t):
    # Definimos todas las líneas posibles de 3 en raya (índices de la lista)
    combinaciones = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8], # Horizontales
        [0, 3, 6], [1, 4, 7], [2, 5, 8], # Verticales
        [0, 4, 8], [2, 4, 6]              # Diagonales
    ]

    for combo in combinaciones:
        # p1, p2, p3 son las posiciones de cada combinación
        p1, p2, p3 = combo

        # Verificamos si las tres posiciones tienen el mismo símbolo
        # y que NO estén vacías (usando el espacio " ")
        if t[p1] == t[p2] == t[p3] != " ":
            return True # ¡Hay un ganador!

    return False # Nadie ha ganado todavía



llamar_funciones()

jugador_actual = "X"
ganador = False

# El bucle se repite mientras no haya ganador Y queden espacios en el Tablero
while not ganador and " " in Tablero:
    mostrar_tablero(Tablero)

    # 1. El jugador actual realiza su jugada
    pedir_jugada(jugador_actual)

    # 2. Verificamos si alguien ganó con esa jugada
    if verificar_ganador(Tablero):
        mostrar_tablero(Tablero)
        print(f"¡FELICIDADES! El jugador {jugador_actual} ha ganado.")
        ganador = True
    else:
        # 3. Si no ganó nadie, verificamos si el tablero está lleno (Empate)
        if " " not in Tablero:
            mostrar_tablero(Tablero)
            print("¡Es un EMPATE! El tablero está lleno.")
        else:
            # 4. Cambiamos de jugador para la siguiente vuelta
            if jugador_actual == "X":
                jugador_actual = "O"
            else:
                jugador_actual = "X"

print("Gracias por jugar.")
```
