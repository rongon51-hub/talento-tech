```python
import math

print("calculadora_Ronald_Gonzalez")

def mostrar_menu():
    print()
    print("Menu escoge tu operacion:")
    print("1. Sumar (+)")
    print("2. Restar (-)")
    print("3. Multiplicar (*)")
    print("4. Dividir (/)")
    print("5. logaritmo (log(10))")
    print("6. Coseno (cos)")
    print("7. Seno (sen)")
    print("8. Tangente (tan)")
    print("9. Exponencial (x^y)")
    print("10. Raiz (x√)")
    print("11. Salir")
    print()

def ingresar_opcion():
    opcion = int(input("ingrese la opcion deseada del menu:"))
    return opcion

def ingresar_numeros():
    numero_uno = float(input("ingrese el primer número de la operacion:"))
    numero_dos = float(input("ingrese el segundo número de la operacion:"))
    return numero_uno, numero_dos

def calcular_suma(numero_uno, numero_dos):
    suma = numero_uno + numero_dos
    return suma

def calcular_resta(numero_uno, numero_dos):
    resta = numero_uno - numero_dos
    return resta

def calcular_multiplicacion(numero_uno, numero_dos):
    multiplicacion = numero_uno * numero_dos
    return multiplicacion

def calcular_division(numero_uno, numero_dos):
    division = numero_uno / numero_dos
    return division

def calcular_logaritmo_base10(numero):
    logaritmo_base10 = math.log10(numero)
    return logaritmo_base10

def calcular_coseno(angulo_radianes):
    coseno = math.cos(angulo_radianes)
    return coseno

def calcular_seno(angulo_radianes):
    seno = math.sin(angulo_radianes)
    return seno

def calcular_tangente(angulo_radianes):
    tangente = math.tan(angulo_radianes)
    return tangente

def calcular_exponencial(numero_uno, numero_dos):
    exponencial = numero_uno**numero_dos
    return exponencial

def calcular_raiz(numero_uno, numero_dos):
    raiz = numero_uno**(1 / numero_dos)
    return raiz

def procesar_calcular_y_mostrar(opcion, numero_uno, numero_dos):
    if opcion == "1" or opcion == "+" or opcion == "suma":
        print("vamos a sumar(+) los dos valores ingresdos:")
        resultado = calcular_suma(numero_uno, numero_dos)
        print("la suma es:", resultado)
    if opcion == "2" or opcion == "-" or opcion == "resta":
        print("vamos a restar(-) los dos valores ingresdos:")
        resultado = calcular_resta(numero_uno, numero_dos)
        print("la resta es:", resultado)
    if opcion == "3" or opcion == "*" or opcion == "multiplicacion":
        print("vamos a multiplicar(*) los dos valores ingresdos:")
        resultado = calcular_multiplicacion(numero_uno, numero_dos)
        print("la multiplicacion es:", resultado)
    if opcion == "4" or opcion == "/" or opcion == "division":
        print("vamos a dividir(/) los dos valores ingresdos:")
        if numero_dos != 0:
            resultado = calcular_division(numero_uno, numero_dos)
            print("la division es:", resultado)
        else:
            print("no es posible dividir entre 0")
    if opcion == "5" or opcion == "log(10)" or opcion == "log":
        print("vamos calcular log(10) de los dos valores ingresdos:")
        if numero_uno > 0:
            resultado1 = calcular_logaritmo_base10(numero_uno)
            print("Resultado1 el log(10) es:", resultado1)
        else:
            print("la funcion logaritmo(x) esta definido para x>0")
        if numero_dos > 0:
            resultado2 = calcular_logaritmo_base10(numero_dos)
            print("Resultado2 el log(10) es:", resultado2)
        else:
            print("la funcion logaritmo(x) esta definido para x>0")

    if opcion == "6" or opcion == "coseno" or opcion == "cos":
        print("vamos calcular coseno de los dos valores ingresdos (en radianes):")
        resultado1 = calcular_coseno(numero_uno)
        resultado2 = calcular_coseno(numero_dos)
        print("Resultado1 el coseno es:", resultado1)
        print("Resultado2 el coseno es:", resultado2)

    if opcion == "7" or opcion == "seno" or opcion == "sen":
        print("vamos calcular seno de los dos valores ingresdos (en radianes):")
        resultado1 = calcular_seno(numero_uno)
        resultado2 = calcular_seno(numero_dos)
        print("Resultado1 el seno es:", resultado1)
        print("Resultado2 el seno es:", resultado2)

    if opcion == "8" or opcion == "tangente" or opcion == "tan":
        print("Vamos a calcular la tangente de los dos valores ingresados (en radianes):")
        if math.isclose(math.cos(numero_uno), 0, abs_tol=1e-15):
            resultado1 = "Indefinido (Coseno es 0)"
        else:
            resultado1 = calcular_tangente(numero_uno)
        if math.isclose(math.cos(numero_dos), 0, abs_tol=1e-15):
            resultado2 = "Indefinido (Coseno es 0)"
        else:
            resultado2 = calcular_tangente(numero_dos)
        print("Resultado1 el tangente es:", resultado1)
        print("Resultado2 el tangente es:", resultado2)

    if opcion == "9" or opcion == "x^y" or opcion == "exponencial":
        print("Vamos a calcular exponencial (x^y) con los dos valores ingresados:")
        try:
            resultado = calcular_exponencial(numero_uno, numero_dos)
            print("El resultado de (numero_uno) elevado a (numero_dos) es:", resultado)
        except OverflowError:
            print("Error: El resultado es demasiado grande para esta calculadora.")
        except ZeroDivisionError:
            print("Error: No se puede elevar  0 a una potencia negativa.")

    if opcion == "10" or opcion == "x√" or opcion == "raiz":
        print("Vamos a calcular la raíz (x√) con los dos valores ingresados:")

        if numero_dos == 0:
            print("Error: El índice de la raíz no puede ser cero.")

        elif numero_uno < 0 and (numero_dos % 2 == 0):
            print("Error: No se puede calcular una raíz par de un número negativo.")

        else:
            resultado = calcular_raiz(numero_uno, numero_dos)
            print("El resultado de la raíz es:", resultado)


def llamar_funciones():
    while True:
        mostrar_menu()
        opcion = input("Ingrese la opcion deseada: ")

        if opcion == '11' or opcion.lower() == "salir":
            print("Hasta Pronto")
            break

        elif opcion > "11":
            print("Opcion no valida")
            break

        try:
            numero_uno, numero_dos = ingresar_numeros()
            procesar_calcular_y_mostrar(opcion, numero_uno, numero_dos)

        except ValueError:
            print("Error: Por favor, ingrese valores numéricos válidos.")

        except Exception as e: # Corrected 'Exepction' to 'Exception'
            print(f"Ocurrió un error inesperado: {e}")

llamar_funciones()
```
